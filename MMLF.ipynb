```python
#@title Evalute
import json
import matplotlib.pyplot as plt

# Đọc dữ liệu từ tệp result.json
result_file_path = "../results/MQR-trec-covid.json"
with open(result_file_path, 'r') as f:
    metrics = json.load(f)

# Trích xuất dữ liệu cho các chỉ số
k_values = [1, 3, 5, 10, 100, 1000]
ndcg = [metrics[f'NDCG@{k}'] for k in k_values]
map_scores = [metrics[f'MAP@{k}'] for k in k_values]
recall = [metrics[f'Recall@{k}'] for k in k_values]
precision = [metrics[f'P@{k}'] for k in k_values]

# Tạo biểu đồ
plt.figure(figsize=(10, 6))
plt.plot(k_values, ndcg, label='NDCG@k', color='#1f77b4', marker='o')
plt.plot(k_values, map_scores, label='MAP@k', color='#ff7f0e', marker='o')
plt.plot(k_values, recall, label='Recall@k', color='#2ca02c', marker='o')
plt.plot(k_values, precision, label='Precision@k', color='#d62728', marker='o')

# Cấu hình biểu đồ
plt.title('Evaluation Metrics for MMLF Pipeline (trec-covid)', fontsize=16)
plt.xlabel('k (Top-k Results)', fontsize=12)
plt.ylabel('Metric Value', fontsize=12)
plt.xscale('log')  # Sử dụng thang log cho trục x để dễ đọc
plt.grid(True, which="both", ls="--")
plt.legend()
plt.ylim(0, 1)  # Giới hạn trục y từ 0 đến 1
plt.show()
```
